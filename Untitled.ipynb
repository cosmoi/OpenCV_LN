```python
    import cv2
    import numpy as np
    from matplotlib import pyplot as plt
    img =  cv2.imread(r"C:\\Users\\cosmoi\\Desktop\\OpenCV_LN\\HGT(1).jpg")
    height, width = img.shape[:2]  # å›¾ç‰‡çš„é«˜åº¦å’Œå®½åº¦

    # (1) ç¼©æ”¾
    fx, fy = 0.6, 0.6
    MAZ = np.float32([[fx, 0, 0], [0, fy, 0]])  # æž„é€ ç¼©æ”¾å˜æ¢çŸ©é˜µ
    imgT1 = cv2.warpAffine(img, MAZ, (width, height))  # ä»¿å°„å˜æ¢, é»‘è‰²å¡«å……
    # (2) å¹³ç§»
    dx, dy = 50, 200  # dx=100 å‘å³åç§»é‡, dy=50 å‘ä¸‹åç§»é‡
    MAT = np.float32([[1, 0, dx], [0, 1, dy]])  # æž„é€ å¹³ç§»å˜æ¢çŸ©é˜µ
    imgT2 = cv2.warpAffine(imgT1, MAT, (width, height), borderValue=(0,255,255))  # å®žçŽ°ä»¿å°„å˜æ¢
    # (3) æ—‹è½¬
    theta = -30 * np.pi / 180  # é€†æ—¶é’ˆæ—‹è½¬ 30Â°
    cosTheta = np.cos(theta)
    sinTheta = np.sin(theta)
    MAR = np.float32([[cosTheta, -sinTheta, 0], [sinTheta, cosTheta, 0]])  # æž„é€ æ—‹è½¬å˜æ¢çŸ©é˜µ
    imgT3 = cv2.warpAffine(imgT2, MAR, (width, height), borderValue=(255,255,0))  # å®žçŽ°ä»¿å°„å˜æ¢
    # (4) æ‰­æ›²
    theta = -30 * np.pi / 180  # é€†æ—¶é’ˆæ‰­å˜ 30Â°
    MAS = np.float32([[1, np.tan(theta), 0], [0, 1, 0]])  # æž„é€ æ‰­å˜å˜æ¢çŸ©é˜µ
    imgT4 = cv2.warpAffine(imgT3, MAS, (width, height), borderValue=(255,0,255))  # å®žçŽ°ä»¿å°„å˜æ¢

    plt.figure(figsize=(9,6))
    plt.subplot(221), plt.axis('off'), plt.title("T1:Zoom")
    plt.imshow(cv2.cvtColor(imgT1, cv2.COLOR_BGR2RGB)),
    plt.subplot(222), plt.axis('off'), plt.title("T2:Translation")
    plt.imshow(cv2.cvtColor(imgT2, cv2.COLOR_BGR2RGB))
    plt.subplot(223), plt.axis('off'), plt.title("T3:Rotation")
    plt.imshow(cv2.cvtColor(imgT3, cv2.COLOR_BGR2RGB))
    plt.subplot(224), plt.axis('off'), plt.title("T4:Shear")
    plt.imshow(cv2.cvtColor(imgT4, cv2.COLOR_BGR2RGB))
    plt.show()
```
